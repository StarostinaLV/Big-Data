```python
import pandas as pd
```

**1. Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы.**

```python
# Читаем файлы:
titles_r = ['user id', 'item id', 'rating', 'timestamp']
movies = pd.read_csv('u.item.csv', sep='|', encoding='latin-1')
ratings = pd.read_csv('u.data.csv', sep = "\t", names = titles_r, header = None)
```

```python
movies.head(2)
```

```python
movies.columns
```

```python
# Последний столбец неинформативен, удалим его:
del movies[';;;;;;;;;;;;;;;;;;;;;;;']
```

```python
# Также оставим только нули и единицы в столбце 'Western':
movies = movies.replace(to_replace =r';*', value='', regex = True)
```

```python
# Приведем заголовки в порядок (чтобы в них не было лишних пробелов):
movies.columns = ['movie_id', 'movie_title', 'release_date', 'video_release_date',
       'IMDb_URL', 'unknown', 'Action', 'Adventure', 'Animation',
       "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
```

```python
movies.head(2)
```

```python
ratings.head(2)
```

**2. Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок.**

```python
# Сгруппируем в датафрейме ratings пользователей по их user id и применим функцию для подсчета рейтинга:
def group_func(data):
  return data.rating.count()
```

```python
user_max = ratings.groupby('user id').apply(group_func)
user_max.max()
```

```python
# Определим user id пользователя, который поставил 737 оценок:
user_max[user_max == 737]
```

Это пользователь под номером 405.

**3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь.**

```python
# Сформируем новый датафрейм из строк, для которых user id равен 405:
ratings_405 = ratings[ratings['user id'] == 405]
ratings_405
```

**4. Добавьте к датафрейму из задания 3 столбцы:**

**По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм принадлежит данному жанру и 0 - если нет
столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей.**

```python
movies_c = movies[['movie_id', 'release_date', 'unknown', 'Action', 'Adventure', 'Animation',
       "Children's", 'Comedy', 'Crime', 'Documentary', 'Drama',
       'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
```

```python
movies_c.head(2)
```

```python
# Одним из признаков будет год, получим его их release_date с помощью функции:
from datetime import datetime

def year_func(data):
  return (str(data))[7:11]
```

```python
movies_c['release_year'] = movies_c['release_date'].apply(year_func)
```

```python
# Столбцес с датой релиза удалим
del movies_c['release_date']
```

```python
# Посмотрим на получившийся датафрейм с частью признаков:
movies_c.head()
```

```python
# Подсчитаем количество оценок:
score_count = ratings.groupby('item id').apply(group_func)
score_count
```

```python
# Подсчитаем суммарную оценку:
score_sum = ratings.groupby('item id').agg(sum)['rating']
score_sum
```

```python
# Из полученных наборов значений соберем датафрейм scores:
scores = pd.DataFrame({'score_count' : score_count, 'score_sum' : score_sum})
scores
```

```python
# Объединим датафрейм scores с датафреймом с фильмами, которые оценивал пользователь 405 (за одно отфильтруем лишних пользователей):
df_405 = ratings_405.merge(scores, how='left', on='item id')
df_405
```

```python
# Датафрейм обогатим признаками, собранными ранее в датафрейм movies_c.
df = df_405.merge(movies_c, how='left', left_on='item id', right_on='movie_id')
df
```

```python
# Столбцы item id и movie_id содержат одни и те же значения, оставим только один из них:
del df['item id']
```

```python
# Столбцец с меткой времени тоже не понадобится:
del df['timestamp']
```

```python
# В результате датафрейм примет следующий вид:
df.head()
```

```python
# Информация о датафрейме df:
df.info()
```

```python
# Данные в столбцах Western и release_year имеют тип 'object', преобразуем их в 'int'.
import numpy as np
```

```python
df = df.astype({'Western':np.int64,'release_year':np.int64})
```

```python
df.info()
```

**9. Средствами спарка вывести среднюю оценку для каждого фильма.**

```python
# Сгруппируем фильмы по номеру (item id) и для каждого посчитаем среднюю оценку:
df_mean_score = ratings.groupby('item id').avg('rating')
df_mean_score.show()
```

**10. В спарке получить 2 датафрейма с 5-ю самыми популярными и самыми непопулярными фильмами (по количеству оценок, либо по самой оценке - на Ваш выбор).**

```python
# Top-5 самых популярных фильмов по количеству оценок:
top5_best = ratings.groupby('item id').count()
top5_best.sort(top5_best['count'].desc()).show(5)
```

```python
# Top-5 самых непопулярных фильмов по количеству оценок:
top5_worst = ratings.groupby('item id').count()
top5_worst.sort(top5_worst['count'].asc()).show(5)
# На самом деле фильмов, которым выставлено по одной оценке, гораздо больше пяти.
```

```python
# Возможно, информативнее будут датафреймы, фильмы в которых отсортированы по среднему рейтингу.
top_avg = df_mean_score.sort(df_mean_score['avg(rating)'].desc())
top_avg.show()
```

```python
# Фильмов с самым высоким средним рейтингом 10 шт.:
top10_avg_best = top_avg
top10_avg_best.show(10)
```

```python
# И 10 фильмов с самым низким средним рейтингом (таких фильмов больше 10-ти):
top10_avg_worst = df_mean_score.sort(df_mean_score['avg(rating)'].asc())
top10_avg_worst.show(10)
```
